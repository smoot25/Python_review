```python
# 붕어빵 철이 돌아왔죠
# 붕어빵(팥) : 2000원, 붕어빵(슈크림) : 2500원, 붕어빵(잡채) : 3000원
# 무인 주문기계를 만들어주세요
# 붕어빵(잡채) 3개, 붕어빵(슈크림) 2개를 주문하면
# "주문이 완료되었습니다. 붕어빵(슈크림) 2개, 붕어빵(잡채) 3개 총 5개 결제 금액은 14000원 입니다."
# 이 출력되도록 코드를 작성해주세요
# 변수를 만들때 스네이크 케이스에 언더바는 2개 / print()문의 포메팅 f-string
```

```python
# 변수
fish_r, fish_s, fish_j = "팥_붕어빵", "슈크림_붕어빵", "잡채_붕어빵"
fish_r_price, fish_s_price, fish_j_price = 2000, 2500, 3000
order_r, order_s, order_j = 2, 2, 1

# 붕어빵 계산
total = order_r + order_s + order_j
r_pay, s_pay, j_pay = fish_r_price * 2, fish_s_price * 2, fish_j_price * 1
total_pay = r_pay + s_pay + j_pay

print(f"붕어빵을 총{total}개 구매하셨습니다.")
print(f"주문하신 붕어빵은 {fish_r}{order_r}개, {fish_s}{order_s}개, {fish_j}{order_j}개 입니다.")
print(f"총 금액은 {total_pay}원 입니다.")
```

```python
# is 와 == 의 차이점
"""
숫자와 문자간에 is연산자 사용 >>>> 참
리스트 or 사전을 is연산자로 비교 >>>> 거짓
리스트 or 사전에서 is연산자 대신 == 로 비교 >>>> 참 (???????)
"""
# 동일성과 동등성의 차이
"""
동일성(identity) : 두 객체의 메모리 주소가 같음.
동등성(equality) : 두 객체의 값이 같음.
is = 동일성을 사용
"""
# 예시
"""
a == b  >>>>  True
a == c  >>>>  True
b == c  >>>>  True
a is b  >>>>  True
a is c  >>>>  False
b is c  >>>>  False
"""

a = [1]
b = a
c = [1]
print(id(a))
print(id(b))
print(id(c))
```

```python
# id 데이터 주소값을 확인하면 a와 b는 주소가 같지만, c는 같지 않다.
# 따라서 is 연산자를 사용할 경우 True가 출력 되지 않는다.
```

```python
# 단락평가

# and 연산자의 왼쪽 피연산자가 False인 경우 오른쪽 피연산자로 넘어가지 않고 확정
# True 일 경우에는 다음 피연산자를 출력.
print(False and True)
print(False and "힘들어요")
print(True and False)
print(True and "힘들어요")
print("\n")
# or 연산자의 왼쪽 피연산자가 True인 경우 오른쪽 피연산자로 넘어가지 않고 확정
# False 일 경우에는 다음 피연산자를 출력.
print(False or True)
print(False or"힘들어요")
print(True or False)
print(True or "힘들어요")
```
